```python
from pprint import pprint
import numpy as np
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
def get_track_waypoints(track_name):
    return np.load("./tracks/%s.npy" % track_name)
    
London_waypoints = get_track_waypoints("London_Loop_Train") ### London loop track
reInv_waypoints = get_track_waypoints("reinvent_base") ### re:invent track

London_waypoints.shape
# reInv_waypoints.shape
```

```python
L_waypts = London_waypoints[:, 0:2]
R_waypts = reInv_waypoints[:, 0:2]
L_waypts[:5]
# R_waypts[:5]
```

```python
L_points_data = pd.DataFrame(L_waypts)
R_points_data = pd.DataFrame(R_waypts)

L_points_data.head()
# R_points_data.head()
```

```python
plt.figure(figsize=(36, 24))

for i in range(len(L_points_data)):
    a = L_points_data.iloc[i][0]
    b = L_points_data.iloc[i][1]
    plt.plot(a, b, 'bo')
    plt.text(a * (1 + 0.01), b * (1 + 0.01) , i, fontsize=12)

plt.xlim((0, 9))
plt.ylim((0, 6))
plt.tight_layout()
plt.savefig("London_waypoints.pdf")
```

```python
plt.figure(figsize=(18, 12))

for i in range(len(R_points_data)):
    a = R_points_data.iloc[i][0]
    b = R_points_data.iloc[i][1]
    plt.plot(a, b, 'bo')
    plt.text(a * (1 + 0.01), b * (1 + 0.01) , i, fontsize=12)

plt.xlim((0, 9))
plt.ylim((0, 6))
plt.tight_layout()
plt.savefig("reinvent_waypoints.pdf")
```
